```python
import re
import matplotlib.pyplot as plt
```

```python
with open("EANN-Full-Training.txt") as f:
    data = f.read()
```

```python
pattern = "Epoch \[(\d+)/\d+\],  Loss: (\d+.\d+), Class Loss: (\d+.\d+), domain loss: (\d+.\d+), Train_Acc: (\d+.\d+),  Validate_Acc: (\d+.\d+)"

data_groups = re.findall(pattern, data)
data_groups[:5]
```

```python
epochs = [int(g[0]) for g in data_groups]
loss = [float(g[1]) for g in data_groups]
class_loss = [float(g[2]) for g in data_groups]
domain_loss = [float(g[3]) for g in data_groups]
train_acc = [float(g[4]) for g in data_groups]
validation_acc = [float(g[5]) for g in data_groups]
```

```python
plt.plot(epochs, train_acc)
plt.plot(epochs, validation_acc)
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Accuracy metrics - EANN Full Weibo Dataset Training over 100 Epochs")
plt.legend(['Training Accuracy', 'Validation Accuracy'])
plt.savefig("eann-full-train-accuracy-metrics.png")
```

```python
plt.plot(epochs, loss)
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss metrics - EANN Full Weibo Dataset Training over 100 Epochs")
plt.savefig("eann-full-train-loss-metrics.png")
# plt.legend(['Loss', 'Class Loss', 'Domain Loss'])
```
